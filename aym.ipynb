```python
#GET BASE KNOWLEDGE 

import pandas as pd
# import pandas to load and process spreadsheet-type data

medical_dataset=pd.read_csv('medical_data.csv') 

# load a medical dataset.

medical_dataset
```

```python
set(medical_dataset['diagnosis'])
```

```python

#SPLIT DB  


from sklearn.model_selection import train_test_split

# use the function to split the data, get 30% for testing and 70% for training.
train_data, test_data = train_test_split(medical_dataset, test_size=0.003, random_state=42)

#print the shapes (rows x columns) of the two datasets
#print(train_data)
print(test_data)
```

```python

#CREATE MODEL


from sklearn.tree import DecisionTreeClassifier

    
# create a Decision Tree Classifier
model=DecisionTreeClassifier(random_state=1)

# get the diagnosis column, to be used as the classification target
train_diagnoses=train_data['diagnosis']
# drop the diagnosis column to get only the symptoms
train_patient_symptoms=train_data.drop(columns=['diagnosis'])
# build a decision tree
model.fit(train_patient_symptoms, train_diagnoses)

# return the trained model
 
```

```python

#TEST DATA


# functions used to evaluate a classifier
from sklearn.metrics import accuracy_score,confusion_matrix

# drop the diagnosis column to get only the symptoms
test_patient_symptoms=test_data.drop(columns=['diagnosis'])
# get the diagnosis column, to be used as the classification target
test_diagnoses=test_data['diagnosis']

 #guess the most likely diagnoses
pred=model.predict(test_patient_symptoms)
print(pred)
#print the achieved accuracy score
accuracy_score(test_diagnoses,pred)
```

```python
confusion_matrix(test_diagnoses,pred)
```

```python
# Ø§Ø³ØªÙŠØ±Ø§Ø¯ Ø§Ù„Ù…ÙƒØªØ¨Ø§Øª
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# ================================
# 1. ØªØ­Ù…ÙŠÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
# ================================
# Ù„Ø§Ø²Ù… ÙŠÙƒÙˆÙ† Ø§Ù„Ù…Ù„Ù medical_data.csv ÙÙŠ Ù†ÙØ³ Ø§Ù„Ù…Ø¬Ù„Ø¯
medical_dataset = pd.read_csv('medical_data.csv')

print("ğŸ“Š Ø­Ø¬Ù… Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª:", medical_dataset.shape)
print("ğŸ“Œ Ø£Ù†ÙˆØ§Ø¹ Ø§Ù„ØªØ´Ø®ÙŠØµ:", set(medical_dataset['diagnosis']))

# ================================
# 2. ØªÙ‚Ø³ÙŠÙ… Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª (70% ØªØ¯Ø±ÙŠØ¨ - 30% Ø§Ø®ØªØ¨Ø§Ø±)
# ================================
train_data, test_data = train_test_split(
    medical_dataset,
    test_size=0.3,   # Ù†Ø³Ø¨Ø© 30% Ù„Ù„Ø§Ø®ØªØ¨Ø§Ø±
    random_state=42
)

# ================================
# 3. Ø¥Ù†Ø´Ø§Ø¡ Ù†Ù…ÙˆØ°Ø¬ Decision Tree
# ================================
model = DecisionTreeClassifier(random_state=1)

# ÙØµÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª: Ø§Ù„Ø£Ø¹Ø±Ø§Ø¶ (features) ÙˆØ§Ù„ØªØ´Ø®ÙŠØµ (target)
train_diagnoses = train_data['diagnosis']
train_patient_symptoms = train_data.drop(columns=['diagnosis'])

# ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
model.fit(train_patient_symptoms, train_diagnoses)

print("\nâœ… ØªÙ… ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø¨Ù†Ø¬Ø§Ø­!")

# ================================
# 4. Ø§Ø®ØªØ¨Ø§Ø± Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
# ================================
test_patient_symptoms = test_data.drop(columns=['diagnosis'])
test_diagnoses = test_data['diagnosis']

# ØªÙ†Ø¨Ø¤Ø§Øª
pred = model.predict(test_patient_symptoms)

print("\nğŸ”® Ø§Ù„ØªÙ†Ø¨Ø¤Ø§Øª:")
print(pred)

# ================================
# 5. ØªÙ‚ÙŠÙŠÙ… Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
# ================================
accuracy = accuracy_score(test_diagnoses, pred)
cm = confusion_matrix(test_diagnoses, pred)

print("\nğŸ¯ Ø¯Ù‚Ø© Ø§Ù„Ù†Ù…ÙˆØ°Ø¬:", accuracy)
print("\nğŸ“Š Ù…ØµÙÙˆÙØ© Ø§Ù„Ø§Ù„ØªØ¨Ø§Ø³ (Confusion Matrix):\n", cm)
```
